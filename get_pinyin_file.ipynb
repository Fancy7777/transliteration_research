```python
def remove_space(word):
    new_word = []
    for ele in word:
        if ele == ' ':
            pass
        else:
            new_word.append(ele)
    return ''.join(new_word)
```

```python
#get in the data
chinese = []
english_phoeneme = []
with open('dataset.txt') as f:
    for line in f:
        temp = line.split('\t')
        chinese.append(remove_space(temp[1]))
        english_phoeneme.append(temp[2])
#english_phoeneme = [phoeneme.split('\n')[0].split() for phoeneme in english_phoeneme]
#chinese = [list(word) for word in chinese]
```

```python
from xpinyin import Pinyin
p = Pinyin()
```

```python
get_pinyin_list = []
for ele in chinese:
    get_pinyin_list.append(p.get_pinyin(ele,' '))
```

```python
with open('pinyin.txt','w') as f1:
    for ele in get_pinyin_list:
        f1.write(ele+'\n')
f1.close()
```
